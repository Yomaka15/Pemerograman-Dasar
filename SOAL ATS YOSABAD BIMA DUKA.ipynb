Program ini digunakan untuk menghitung efisiensi turbin dengan cara membandingkan energi yang masuk dan energi yang keluar.

Fungsi hitung_efisiensi(input_energi, output_energi):

Fungsi ini menerima dua nilai: energi yang masuk (input_energi) dan energi yang keluar (output_energi).

  Rumus efisiensinya adalah output_energi / input_energi. Jadi, program akan membagi energi yang keluar dengan energi yang masuk.

- Cara Kerja:

  input energi adalah 500 dan output energi adalah 400. Program akan menghitung efisiensinya dengan rumus tersebut.

- Hasil:

  Program akan menampilkan hasil efisiensi turbin, yang dalam kasus ini adalah 0.8 (artinya 80%).

- Masalah yang diselesaikan:

  Program ini membantu menghitung seberapa efisien   turbin bekerja berdasarkan energi yang masuk dan keluar.

```python
def hitung_efisiensi(input_energi, output_energi):
    """Menghitung efisiensi turbin berdasarkan input dan output energi."""
    efisiensi = output_energi / input_energi
    return efisiensi

# Memanggil fungsi dengan input 500 dan output 400
input_energi = 500
output_energi = 400
efisiensi = hitung_efisiensi(input_energi, output_energi)

print(f"Efisiensi turbin adalah: {efisiensi:.2f}")
```

Program ini memeriksa apakah katup pengaman bisa dibuka atau tetap tertutup berdasarkan tekanan uap dan status manual override.

Fungsi cek_katup_pengaman(tekanan_uap, status_manual_override):

Fungsi ini menerima dua input: tekanan uap (dalam psi) dan status manual override (apakah manual override aktif atau tidak).

- Katup akan terbuka jika:

  Tekanan uap lebih dari 150 psi dan

  Manual override dalam keadaan off (False).

  Jika salah satu syarat tidak terpenuhi, katup akan tetap tertutup.

- Cara Kerja:

  tekanan uap adalah 150 psi dan manual override adalah False. Karena tekanan uap tidak lebih dari 150 psi, katup tetap tertutup.

- Hasil:

  Program akan menampilkan "Katup Tertutup" karena tekanan uap tidak cukup tinggi.

- Masalah yang diselesaikan:
  Program ini membantu mengecek apakah katup pengaman  bisa dibuka atau tetap tertutup berdasarkan kondisi yang diberikan.

```python
def cek_katup_pengaman(tekanan_uap, status_manual_override):
    """Memeriksa kondisi untuk membuka katup pengaman."""
    if tekanan_uap > 150 and not status_manual_override:
        print("Katup Terbuka")
    else:
        print("Katup Tertutup")

# Contoh pemanggilan fungsi
tekanan_uap = 160.0  # dalam psi
status_manual_override = False

cek_katup_pengaman(tekanan_uap, status_manual_override)
```

```python
import time
def hitung_energi(daya_panel,lama_penyinaran,efisiensi):
  """
  Menghitung energi listrik (Wh) yang di hasilkan panel surya.
  Rumus: Energi = Daya Panel * Lama Penyinaran * Efesiensi
  """
  energi = daya_panel * lama_penyinaran * efisiensi
  return energi
def kategori_penyinaran(jam):
  if jam < 2:
    return "Mendung"
  elif jam <5:
    return "Cerah sebagian"
  else:
    return "Sangat cerah"
print("=== SIMULASI ENERGI SURYA ===")
print("Program ini menghitung total energi listrik dari panel surya dalam 7 hari.\n ")

daya_panel = float(input("Masukkan daya panel surya (watt): "))
efisiensi = float(input("masukan efisiensi panel (contoh 0.8 unntuk 80%)"))
total_energi = 0
hari = 1
while hari <= 7:
  print(f"\nHari ke-{hari}---")
  lama_penyinaran = float(input("Masukkan lama penyinaran (jam): "))
  energi_harian = hitung_energi(daya_panel,lama_penyinaran,efisiensi)
  total_energi += energi_harian
  hari += 1 # Increment hari to avoid infinite loop
print("\n=== HASIL AKHIR ===")
print(f"Total energi listrik selama 7 hari:{total_energi:.2f} Wh")
if total_energi < 2000:
  print("Efisiensi sistem masih rendah , pertimbangkan untuk menambah panel")
elif total_energi < 5000:
  print("Kinerja panel cukup baik. ")
else:
  print("Sistem panel surya sangat efisiensi dan optimal!")
print("\nMenyiapkan laporan",end= "")
for i in range(5):
  print(".",end="",flush=True)
  time.sleep(0.3)
print("\nLaporan energi selesai di buat!")
```
